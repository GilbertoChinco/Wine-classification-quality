## Data visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Import from the previous file *01_Data_Cleaning_wine* the dataset .csv 

```python
df = pd.read_csv("winequality-red-white.csv")
df.head()
```

The variables for the dataset:

Input variables (based on physicochemical tests):
1. fixed acidity
2. volatile acidity
3. citric acid
4. residual sugar
5. chlorides
6. free sulfur dioxide
7. total sulfur dioxide
8. density
9. pH
10. sulphates
11. alcohol

Output variable (based on sensory data):
12. quality (score between 0 and 10)
13. Class --> **0 : "red-wine" ||  1 : "white-white"** 

```python
df.columns
```

Looking the *acid* variables **'fixed acidity', 'volatile acidity', 'citric acid'** divided by **class**

```python
sns.pairplot(df[['fixed acidity', 'volatile acidity', 'citric acid', 'class']], hue = "class")
```

```python
fig, axes = plt.subplots(1, 3, sharex = True, figsize = (16, 7))
sns.boxplot(x = "class", y = "fixed acidity", data = df, ax=axes[0])
axes[0].set(title = "Comparision fixed acidity")
sns.boxplot(x = "class", y = "volatile acidity", data = df, ax=axes[1])
axes[1].set(title = "Comparision volatile acidity")
sns.boxplot(x = "class", y = "citric acid", data = df, ax=axes[2])
axes[2].set(title = "Comparision citric acid")
```

Looking the variables **residual sugar, chlorides, free sulfur dioxide, total sulfur dioxide** divided by **class**

```python
sns.pairplot(df[[ 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', "class"]], hue = "class")
```

```python
fig, axes = plt.subplots(1, 4, sharex = True, figsize = (20, 7))
sns.boxplot(x = "class", y = "residual sugar", data = df, ax=axes[0])
axes[0].set(title = "Comparision residual sugar")
sns.boxplot(x = "class", y = "chlorides", data = df, ax=axes[1])
axes[1].set(title = "Comparision chlorides")
sns.boxplot(x = "class", y = "free sulfur dioxide", data = df, ax=axes[2])
axes[2].set(title = "Comparision free sulfur dioxide")
sns.boxplot(x = "class", y = "total sulfur dioxide", data = df, ax=axes[3])
axes[3].set(title = "Comparision total sulfur dioxide")
```

Looking the variables **density, pH, sulphates, alcohol** divided by **class**

```python
sns.pairplot(df[['density', 'pH', 'sulphates', 'alcohol', 'class']], hue = "class")
```

```python
fig, axes = plt.subplots(1, 4, sharex = True, figsize = (20, 7))
sns.boxplot(x = "class", y = "density", data = df, ax=axes[0])
axes[0].set(title = "Comparision density")
sns.boxplot(x = "class", y = "pH", data = df, ax=axes[1])
axes[1].set(title = "Comparision pH")
sns.boxplot(x = "class", y = "sulphates", data = df, ax=axes[2])
axes[2].set(title = "Comparision sulphates")
sns.boxplot(x = "class", y = "alcohol", data = df, ax=axes[3])
axes[3].set(title = "Comparision alcohol")
```

**Correlation between variables**

```python
df[['fixed acidity', 'volatile acidity', 'citric acid',
       'residual sugar', 'chlorides', 'free sulfur dioxide',
       'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol',
       'quality', 'class']].corr()
```

```python
fig, ax = plt.subplots(figsize = (10, 10)) 
sns.heatmap(df[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide',
       'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol', 'quality', 'class']].corr(), annot = True, ax = ax)
```
